>#### Exploratory Analysis on Presidents' Heights

This report delves into the heights of presidents to uncover trends in their stature. It also poses an intriguing question: "What is the probability of having a president of a certain height?" This query aims to understand the likelihood of specific heights among global presidents, adding a predictive dimension to our exploration

>#### Importing and Loading Data into the df Variable

```python
import pandas as pd
df = pd.read_csv("/president_heights.csv")
```

> #### Printing the first 5 rows of the dataset

```python
df.head()
```

> #### Printing relevant information about the dataset

```python
df.info()
```

> #### Transformation of the height(cm) column into a Numpy array in the variable height.

```python
import numpy as np
height = np.array(df["height(cm)"])
height
```

>#### Statistical Measurements (Mean, Std, Min, and Max)

This analysis explores the heights of presidents

Key Statistics:

Average Height: 179.74 cm
Standard Deviation: 6.93 cm
Minimum Height: 163 cm
Maximum Height: 193 cm

A smaller standard deviation indicates that the data points are closely clustered around the mean, while a larger standard deviation indicates that the data points are more spread out.

In this case, the standard deviation is approximately 6.93 cm. This means that, on average, the heights of presidents in the dataset deviate from the mean height of 179.74 cm by about 6.93 cm.

```python
print(f'Average height: {height.mean()}' )
print(f'Standard Deviation: {height.std()}' )
print(f'Min height: {height.min()}' )
print(f'Max height: {height.max()}' )
```

> #### Quartile Measurements and Data Distribution Check (25%, Median, 75%)

The height metrics of presidents can be better understood by analysing their distribution. The 25th percentile height of 174.25 cm indicates that 25% of the presidents were shorter than this height, while the 75th percentile height of 183.0 cm suggests that 75% of the presidents were shorter than this height. The median height of 182.0 cm represents the middle value in the dataset and is a good indicator of the central tendency of president heights. By examining these metrics, we can gain a more comprehensive understanding of the height distribution among presidents.

```python
print(f'The 25th percentile height:  {np.percentile(height,25)}')
print(f'The Median height:  {np.median(height)}')
print(f'The 75th percentile height: {np.percentile(height,75)}')
```

> #### Histogram Plotting

This code imports two Python libraries: matplotlib.pyplot and seaborn.
Matplotlib.pyplot is used for creating a variety of static, animated, and interactive visualizations in Python.
Seaborn is a statistical data visualization library built on top of matplotlib.
The sns.set() command enhances the visual aesthetics of matplotlib plots using seaborn's default style settings.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

When we look at the graph below, we can see that they range from 165 cm to 190 cm. The most common height range among the presidents is between 180 cm and 185 cm. There are fewer presidents who measure below 170 and an average line just before 180. By examining this graph, we can understand the height distribution among presidents.


```python
avg_height = height.mean()
plt.hist(height)
plt.axvline(avg_height, color='red', linestyle='dashed', linewidth=2)
plt.savefig('saved_figure.png')
```

This code creates a histogram of the data array height using matplotlib.pyplot (plt). It then adds a dashed red vertical line at the position avg_height, representing the average height, with a specified line width. Finally, the plot is saved to a file named 'saved_figure.png'.
